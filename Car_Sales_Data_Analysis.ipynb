```python
import pandas as pd


df = pd.read_csv("Car_sales.csv")

print(df.head())
```

```python
print(df.columns)
print(df.info())
```

```python
sales_by_model = df.groupby("Model")["Sales_in_thousands"].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
sales_by_model.plot(kind="pie", autopct='%1.1f%%', startangle=140, colormap="Set3")
plt.title("Sales Distribution by Car Model", fontsize=16)
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sales_by_model.plot(kind="pie", autopct='%1.1f%%', startangle=140, colormap="Set3")
plt.title("Sales Distribution by Car Model", fontsize=16)
plt.ylabel("")
plt.show()
```

```python
import plotly.express as px
fig_bar = px.bar(
    sales_by_model.reset_index(),
    x="Model",
    y="Sales_in_thousands",
    title="Top Selling Car Models",
    color="Sales_in_thousands",
    color_continuous_scale="Blues"
)
fig_bar.show()
```

```python
fig_pie = px.pie(
    sales_by_model.reset_index(),
    names="Model",
    values="Sales_in_thousands",
    title="Sales Distribution by Car Model"
)
fig_pie.show()
```
